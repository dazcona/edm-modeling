```python
df_web = pd.read_csv('../../data/raw/web_data.csv')
```

```python
df_web['date'] = pd.to_datetime(df_web['date'], format='%y%m%d-%H:%M.%S')
```

```python
df_web['date'].dt.date.value_counts().plot(figsize=(16, 8), color='b', fontsize=15)
df_programs['date'].dt.date.value_counts().plot(figsize=(16, 8), color='darkcyan', fontsize=15)
```

```python
# Second Y-axis
# http://matplotlib.org/examples/api/two_scales.html
fig, ax1 = plt.subplots(figsize=(18, 10))
plt.grid()
# PLOT FIRST
df_web['date'].dt.date.value_counts().plot(figsize=(16, 8), color='b', fontsize=15, ax=ax1)
# ax1.set_xlabel("Lab dates", fontsize=30)
ax1.set_ylabel("Web requests", color='b', fontsize=30)
# Make the y-axis label and tick labels match the line color.
for tl in ax1.get_yticklabels():
    tl.set_color('b')
ax2 = ax1.twinx()
# PLOT SECOND
df_programs['date'].dt.date.value_counts().plot(figsize=(16, 8), color='darkcyan', fontsize=15, ax=ax2)
ax2.set_ylabel('Programs submitted', color='darkcyan', fontsize=30)
for tl in ax2.get_yticklabels():
    tl.set_color("darkcyan")
# rotate and align the tick labels so they look better
fig.autofmt_xdate()
filename = '../../figures/exploring_Programming_Web_dates.png'
plt.savefig(filename, format='png', dpi=100)
```

### Academic year: 2016/2017

```python
from datetime import datetime
```

```python
start = datetime(year=2016, month=9, day=1)
end = datetime(year=2017, month=2, day=1)
```

```python
def plot_academic_year(start, end, title):
    # Second Y-axis
    # http://matplotlib.org/examples/api/two_scales.html
    fig, ax1 = plt.subplots(figsize=(18, 10))
    plt.grid()
    # PLOT FIRST
    df_web['date'][
        (df_web['date'] > start) & 
        (df_web['date'] < end)].dt.date.value_counts().plot(figsize=(16, 8), color='b', fontsize=15, ax=ax1)
    # ax1.set_xlabel("Lab dates", fontsize=30)
    ax1.set_ylabel("Web requests", color='b', fontsize=30)
    # Make the y-axis label and tick labels match the line color.
    for tl in ax1.get_yticklabels():
        tl.set_color('b')
    ax2 = ax1.twinx()
    # PLOT SECOND
    df_programs['date'][
        (df_programs['date'] > start) & 
        (df_programs['date'] < end)].dt.date.value_counts().plot(figsize=(16, 8), color='darkred', fontsize=15, ax=ax2)
    ax2.set_ylabel('Programs submitted', color='darkred', fontsize=30)
    for tl in ax2.get_yticklabels():
        tl.set_color("darkred")
    # rotate and align the tick labels so they look better
    fig.autofmt_xdate()
    filename = '../../figures/{}.png'.format(title)
    plt.savefig(filename, format='png', dpi=100)
```

```python
plot_academic_year(start, end, 'exploring_Programming_Web_dates_2016_2017')
```

### Academic year: 2017/2018

```python
start = datetime(year=2017, month=9, day=1)
end = datetime(year=2018, month=2, day=1)
```

```python
plot_academic_year(start, end, 'exploring_Programming_Web_dates_2017_2018')
```
